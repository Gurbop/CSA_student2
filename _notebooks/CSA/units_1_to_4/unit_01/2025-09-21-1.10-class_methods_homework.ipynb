# Homework: Ultimate Battle

## Scenario

You are programming a game where 2 objects of your choosing (e.g., Robots, Dinosaurs, People) battle. Each object has health and power.

---

## Instructions

1. **Create a class** representing your object with:

   **Instance variables:**  
   - `String name`  
   - `int power`  
   - `int health`  
   - `MAKE YOUR OWN`

   **Static variables:**
   - `double fightDuration`

   **Instance methods:**  
   - `void attack()`
   - `void printStatus()`
   - `MAKE YOUR OWN`

   **Class methods:**  
   - `static int strongerFighter()`
   - `static void beginBattle()`
   - `MAKE YOUR OWN`

2. **In `main`:**  
   - Create two objects.  
   - Use instance methods to attack and print status.  
   - Use static methods to compare, print a fact, and start the battle.

```java
// Code Here
public class Wizard {
    // Instance variables
    String name;
    int power;
    int health;
    int mana;  // unique variable to boost attack

    // Static variable
    static double fightDuration = 0;

    // Constructor
    public Wizard(String name, int power, int health, int mana) {
        this.name = name;
        this.power = power;
        this.health = health;
        this.mana = mana;
    }

    // Instance method: attack another wizard
    public void attack(Wizard enemy) {
        int damage = power + (mana / 2); // mana boosts attack
        System.out.println(name + " attacks " + enemy.name + " for " + damage + " damage!");
        enemy.health -= damage;
        if (enemy.health < 0) enemy.health = 0;
        fightDuration += 1;  // each attack adds 1 second to fight duration
    }

    // Instance method: print current status
    public void printStatus() {
        System.out.println(name + " - Health: " + health + ", Power: " + power + ", Mana: " + mana);
    }

    // Class method: compare who is stronger by power
    public static int strongerFighter(Wizard w1, Wizard w2) {
        if (w1.power > w2.power) return 1;
        else if (w2.power > w1.power) return 2;
        else return 0; // tie
    }

    // Class method: begin battle between two wizards
    public static void beginBattle(Wizard w1, Wizard w2) {
        System.out.println("The battle begins between " + w1.name + " and " + w2.name + "!");
        while (w1.health > 0 && w2.health > 0) {
            w1.attack(w2);
            if (w2.health <= 0) break;
            w2.attack(w1);
        }
        System.out.println("Battle over in " + fightDuration + " seconds.");
        if (w1.health > 0) System.out.println(w1.name + " wins!");
        else System.out.println(w2.name + " wins!");
    }

    // Main method
    public static void main(String[] args) {
        Wizard merlin = new Wizard("Merlin", 25, 100, 40);
        Wizard morgana = new Wizard("Morgana", 30, 90, 35);

        merlin.printStatus();
        morgana.printStatus();

        int stronger = Wizard.strongerFighter(merlin, morgana);
        if (stronger == 1) System.out.println(merlin.name + " is the stronger fighter.");
        else if (stronger == 2) System.out.println(morgana.name + " is the stronger fighter.");
        else System.out.println("Both fighters are equally strong.");

        Wizard.beginBattle(merlin, morgana);

        merlin.printStatus();
        morgana.printStatus();
    }
}
```
